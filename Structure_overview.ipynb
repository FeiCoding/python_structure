# Insert Time Measurement
In this part, we will test insert data into different data structures to test the average time for each and then print out a graph to demonstrate and compare the results between each other.

### Format/Insert data
We will use 1 million data to insert to the datastructures showed below:
1. List
2. Tree
3. Set
4. Dictionary
5. LinkedList

##### Due to the similar structure of stack and list, the data insert time should share the similar time.

```python
DATA_SIZE = 1000000
DATA_RANGE = 1000000
INSERT_LOOP_TIME = 10
runner = Runner(DATA_SIZE, DATA_RANGE)
data_list, list_form_time, data_tree, tree_form_time, data_set, set_form_time, data_dict, dict_form_time, data_linked_list, linked_list_form_time, data_sort_list, sort_list_form_time = runner.formate_data(INSERT_LOOP_TIME)
time_insert = []
name = []
table_plot = [[]]
```

```python
time_insert = [list_form_time, sort_list_form_time, tree_form_time, dict_form_time, dict_form_time, linked_list_form_time]
name = ['List', 'Sorted List', 'Tree', 'Set', 'Dictionary', 'Linked List']
```

```python
plot = Plotter()
plot.diagram_plot(time_insert, name, operation_name = "Formate Time")
```

## Search Problem
We will use a huge dataset to test different data structures

### Define data holder and a random index list between 0 and 9999 for searching targets 

```python
time_search = []
SEARCH_LOOP_TIME = 10
targets = [random.randint(0,DATA_SIZE) for i in range(SEARCH_LOOP_TIME)]
print(targets)
```

##### List Searching can only rely on iterating the whole list to compare each item 

```python
list_search_time = runner.search("list", SEARCH_LOOP_TIME, targets, data_list)
time_search.append(list_search_time)
print(list_search_time)
```

### Sorted List Searching

```python
sorted_list_search_time =  runner.search("sort_list", SEARCH_LOOP_TIME, targets, data_sort_list)
time_search.append(sorted_list_search_time)
print(sorted_list_search_time)
```

### Tree Searching 

```python
tree_search_time =  runner.search("tree", SEARCH_LOOP_TIME, targets, data_tree)
time_search.append(tree_search_time)
print(tree_search_time)
```

#### Analysis: Here we use round, so 0 means the search time is too short to record

### Set Searching

```python
set_search_time =  runner.search("set", SEARCH_LOOP_TIME, targets, data_set)
time_search.append(set_search_time)
print(set_search_time)
```

### Dictionary Searching

```python
dict_search_time =  runner.search("dictionary", SEARCH_LOOP_TIME, targets, data_dict)
time_search.append(dict_search_time )
print(dict_search_time)
```

### LinkedList Searching

```python
linkedlist_search_time = runner.search("linked_list", SEARCH_LOOP_TIME, targets, data_linked_list)
time_search.append(linkedlist_search_time)
print(linkedlist_search_time)
```

### Plot the search time diagram

```python
print(len(time_search))
plot.diagram_plot(time_search, name, operation_name = "Search Time")
```
