# Insert Time Measurement
In this part, we will test insert data into different data structures to test the average time for each and then print out a graph to demonstrate and compare the results between each other.

### Format data
We will use 1 million data to insert to the datastructures showed below:
1. List
2. Tree
3. Set
4. Dictionary
5. LinkedList

```python
DATA_SIZE = 100000000
DATA_RANGE = 100000000
INSERT_LOOP_TIME = 10
data = random.sample(range(0, DATA_RANGE), DATA_SIZE)
time_insert = []
name = []
table_plot = [[]]
```

### Insert data into list structure 

##### Due to the similar structure of stack and list, the data insert time should share the similar time.

```python
count_time = 0
data_list = []
for count in range(INSERT_LOOP_TIME):
    start_time = timeit.default_timer()
    for num in data:
        data_list.append(num)
    count_time += timeit.default_timer() - start_time

list_insert_time = round(count_time / INSERT_LOOP_TIME, 4)
print(list_insert_time)
time_insert.append(list_insert_time)
name.append('List')
```

### Insert data into tree structure

```python
count_time = 0
data_tree = Tree()
for count in range(INSERT_LOOP_TIME):
    start_time = timeit.default_timer()
    for num in data:
        data_tree.insert(num)
    count_time += timeit.default_timer() - start_time
    
tree_insert_time = round(count_time / INSERT_LOOP_TIME, 4)
print(tree_insert_time)
time_insert.append(tree_insert_time)
name.append('Tree')
```

### Insert data into set structure 

```python
count_time = 0
data_set = set()
for count in range(INSERT_LOOP_TIME):
    start_time = timeit.default_timer()
    for num in data:
        data_set.add(num)
    count_time += timeit.default_timer() - start_time

set_insert_time = round(count_time / INSERT_LOOP_TIME, 4)
print(set_insert_time)
time_insert.append(set_insert_time)
name.append('Set')
```

### Insert data into dictionary structure 

```python
count_time = 0
data_dict = {}
for count in range(INSERT_LOOP_TIME):
    start_time = timeit.default_timer()
    for num in data:
        data_dict[num] = num
    count_time += timeit.default_timer() - start_time

dict_insert_time = round(count_time / INSERT_LOOP_TIME, 4)
print(dict_insert_time)
time_insert.append(dict_insert_time)
name.append('Dictionary')
```

### Insert data into single linked list data structure 

```python
count_time = 0
data_linked_list = LinkedList()
for count in range(INSERT_LOOP_TIME):
    start_time = timeit.default_timer()
    for num in data:
        data_linked_list.addNode(num)
    count_time += timeit.default_timer() - start_time

linked_list_insert_time = round(count_time / INSERT_LOOP_TIME, 4)
print(linked_list_insert_time)
time_insert.append(linked_list_insert_time)
name.append('Linked List')
```

### Plot the Diagram 

```python
plot = Plotter()
plot.diagram_plot(time_insert, name, operation_name = "Insert Time")
```

## Search Problem
We will use a huge dataset to test different data structures

### Define data holder and a random index list between 0 and 9999 for searching targets 

```python
time_search = []
SEARCH_LOOP_TIME = 10
targets = [random.randint(0,DATA_SIZE) for i in range(SEARCH_LOOP_TIME)]
print(targets)
```

##### List Searching can only rely on iterating the whole list to compare each item 

```python
count_time = 0
for target in targets:
    start_time = timeit.default_timer()
    for i in range(len(data_list)):
        if data_list[i] == data[target]:
            count_time += round((timeit.default_timer() - start_time) * 1000)
            break
list_search_time = count_time / SEARCH_LOOP_TIME 
time_search.append(list_search_time)
print(list_search_time)
```

### Tree Searching 

```python
count_time = 0
for target in targets:
    start_time = timeit.default_timer()
    data_tree.search(data[target])
    count_time += round((timeit.default_timer() - start_time) * 1000)
tree_search_time =  count_time / SEARCH_LOOP_TIME
time_search.append(tree_search_time)
print(tree_search_time)
```

#### Analysis: Here we use round, so 0 means the search time is too short to record

### Set Searching

```python
count_time = 0
for target in targets:
    start_time = timeit.default_timer()
    data[target] in data_set
    count_time += round((timeit.default_timer() - start_time) * 1000)
set_search_time = count_time / SEARCH_LOOP_TIME
time_search.append(set_search_time)
print(set_search_time)
```

### Dictionary Searching

```python
count_time = 0
for target in targets:
    start_time = timeit.default_timer()
    if data[target] in data_dict:
        pass
    count_time += round((timeit.default_timer() - start_time) * 1000)
dict_search_time = count_time / SEARCH_LOOP_TIME
time_search.append(dict_search_time )
print(dict_search_time)
```

### LinkedList Searching

```python
count_time = 0
for target in targets:
    start_time = timeit.default_timer()
    data_linked_list.search(data[target])
    count_time += round((timeit.default_timer() - start_time) * 1000)
linkedlist_search_time = count_time / SEARCH_LOOP_TIME
time_search.append(linkedlist_search_time)
print(linkedlist_search_time)
```

### Plot the search time diagram

```python
print(len(time_search))
plot.diagram_plot(time_search, name, operation_name = "Search Time")
```
