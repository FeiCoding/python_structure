### Different Kinds of Tree Structure

#### This section will compare the insert time on sorted list data of different Tree structures such as 
1. Normal Binary Search Tree
2. AVL Tree 
3. Red-Black Tree 


```python
from RBTree import * 
from AVLTree import *
```

```python
data_tree = random.sample(range(0, 100000), 1000)
data_tree.sort()
tree = Tree()
start_time = timeit.default_timer()
for num in data_tree:
    tree.insert(num)
normal_tree_time = round(timeit.default_timer() - start_time, 3)
```

```python
print(normal_tree_time)
```

```python
avl_tree = AVLTree()
start_time = timeit.default_timer()
for num in data_tree:
    avl_tree.insert(num)
avl_tree_time = round(timeit.default_timer() - start_time, 3)
print(avl_tree_time)
```

```python
rb_tree = EmptyRedBlackTree()
start_time = timeit.default_timer()
for num in data_tree:
    rb_tree.insert(num)
rb_tree_time = round(timeit.default_timer() - start_time, 3)
print(rb_tree_time)
```

```python
name = ["Binary Search Tree", "AVL Tree", "Read-Black Tree"]
time_tree = [normal_tree_time, avl_tree_time, rb_tree_time]
plot.diagram_plot(time_tree, name, "Inserting Time")
```
