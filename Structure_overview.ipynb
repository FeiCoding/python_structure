# Insert Problem 
In this part, we will test insert data into different data structures to test the average time for each and then print out a graph to demonstrate and compare the results between each other.

### Format data
We will use 1 million data to insert to the datastructures showed below:
1. List
2. Tree
3. Set
4. Dictionary
5. LinkedList

```python
DATA_SIZE = 1000000
DATA_RANGE = 10000
INSERT_TIME = 10
data = [random.randint(0, DATA_RANGE) for r in range(DATA_SIZE)]
time_insert = []
name = []
table_plot = [[]]
```

### Insert data into list structure 

##### Due to the similar structure of stack and list, the data insert time should share the similar time.

```python
count_time = 0
data_list = []
for count in range(INSERT_TIME):
    start_time = time.clock()
    for num in data:
        data_list.append(num)
    count_time += time.clock() - start_time

list_insert_time = round(count_time / INSERT_TIME, 4)
print(list_insert_time)
time_insert.append(list_insert_time)
name.append('List')
```

### Insert data into tree structure

```python
count_time = 0
data_tree = Tree()
for count in range(INSERT_TIME):
    start_time = time.clock()
    for num in data:
        data_tree.insert(num)
    count_time += time.clock() - start_time
    
tree_insert_time = round(count_time / INSERT_TIME, 4)
print(tree_insert_time)
time_insert.append(tree_insert_time)
name.append('Tree')
```

### Insert data into set structure 

```python
count_time = 0
data_set = set()
for count in range(INSERT_TIME):
    start_time = time.clock()
    for num in data:
        data_set.add(num)
    count_time += time.clock() - start_time

set_insert_time = round(count_time / INSERT_TIME, 4)
print(set_insert_time)
time_insert.append(set_insert_time)
name.append('Set')
```

### Insert data into dictionary structure 

```python
count_time = 0
data_dict = {}
for count in range(INSERT_TIME):
    start_time = time.clock()
    for num in data:
        data_dict[num] = num
    count_time += time.clock() - start_time

dict_insert_time = round(count_time / INSERT_TIME, 5)
print(dict_insert_time)
time_insert.append(dict_insert_time)
name.append('Dictionary')
```

### Insert data into single linked list data structure 

```python
count_time = 0
data_linked_list = LinkedList()
for count in range(INSERT_TIME):
    start_time = time.clock()
    for num in data:
        data_linked_list.addNode(num)
    count_time += time.clock() - start_time

linked_list_insert_time = round(count_time / INSERT_TIME, 4)
print(linked_list_insert_time)
time_insert.append(linked_list_insert_time)
name.append('Linked List')
```

### Plot the Diagram 

```python
fig, ax = plt.subplots()
print(time_insert)
index = range(len(name))

print(index)
ax.bar(index, time_insert)
ax.set_xticks(range(len(name)))
ax.set_xticklabels(name)
ax.set_xlabel('Data Structure Name', weight='semibold')
ax.set_ylabel('Time', weight='semibold')
ax.set_title("Insert Time of Different Data Structure", fontsize=15)
for a,b in zip(index, time_insert):
    plt.text(a - 0.3, b+0.0003, str(b))
plt.show()

    
table_plot = [['Name', 'Insert Time']]
for i in range(len(name)):
    table_plot.append([name[i], time_insert[i]])
display(HTML(tabulate.tabulate(table_plot, tablefmt='html')))
```
