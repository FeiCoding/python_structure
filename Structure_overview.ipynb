# Normal List

# Stack Structure


```python
#use list as a stack
stack = [3, 4, 5]
listviz(stack)
```

```python
#push item into the end of the list
stack.append(6)
stack.append(7)
listviz(stack)
```

```python
#pop item out of the list
stack.pop()
listviz(stack)
```

# Double List(Table/Matrix)

```python
#table/matrix structure 
table=[
    ['Year', 'Name', 'Grade', 'Class Number', 'Score'],
    [2019,'Jack','First year',3,82],
    [2019,'Lily','First year',3,76],
    [2019,'John','First year',3,90],
    [2019,'Will','First year',3,56],
    [2019,'Shawn','First year',3,78],
]
objviz(table)
```

# Set Structure 

```python
#set structrue
basket = {"apple", "banana", "pear", "apple"}
objviz(basket) #set will remove the duplicated item
```

# Queue Structure

```python
q = queue.Queue()
for i in range(5):
    q.put(i)
objviz(q)
```
